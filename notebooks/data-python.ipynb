```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Generate some 2D sample data
X_train = np.random.rand(50, 2) 
y_train = 2 * X_train[:, 0] - 3 * X_train[:, 1] + 4 + 0.1 * np.random.randn(50)

# Create a LinearRegression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Print the coefficients
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")

# Use the model to make predictions on a set of test data
X_test = [[0.5, 0.5], [0, 0], [1, 1]]
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot for the training data and model predictions

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot training data
ax.scatter(X_train[:, 0], X_train[:, 1], y_train, c='blue', marker='o', label='Training data')

# Create a mesh for the model's prediction surface
x0_range = np.linspace(X_train[:, 0].min(), X_train[:, 0].max(), 10)
x1_range = np.linspace(X_train[:, 1].min(), X_train[:, 1].max(), 10)
x0_mesh, x1_mesh = np.meshgrid(x0_range, x1_range)
y_mesh = model.predict(np.c_[x0_mesh.ravel(), x1_mesh.ravel()]).reshape(x0_mesh.shape)

# Plot the fitted plane
ax.plot_surface(x0_mesh, x1_mesh, y_mesh, alpha=0.5, color='red', label='Fitted model')

ax.set_xlabel('X₀')
ax.set_ylabel('X₁')
ax.set_zlabel('y')
ax.set_title('Linear Regression Model: Training Data and Fitted Plane')
ax.legend()

plt.show()
```
